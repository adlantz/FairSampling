```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
import numpy as np

DB_PATH = '../database/fair_sampling.db'
```

```python
# Establish a connection to the SQLite database
conn = sqlite3.connect(DB_PATH)

# Define your SQL query

query  = """

SELECT post_anneal_supp_ratio, disc_fair_sampling, disc_post_anneal, disc_fair_sampling - disc_post_anneal as disc_diff  FROM instances_N12
WHERE  degeneracy > 4 and post_anneal_supp_ratio > 0.98;

"""

# Execute the query and read the data into a pandas DataFrame
df = pd.read_sql_query(query, conn)

# Close the database connection
conn.close()
```

```python
df
```

```python
N = 8

fair_discs = df["disc_fair_sampling"]
pa_discs = df["disc_post_anneal"]

bins = [0.5+i for i in range(N+1)]
print(bins)
y_limits = (0, len(fair_discs)) 
x_limits = (0,N+1)


# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Create the first histogram
ax1.hist(fair_discs, bins=bins, edgecolor='black')
ax1.set_title('Fair')
ax1.set_xlabel('Disconnectivity')
ax1.set_ylabel('Frequency')
ax1.set_ylim(y_limits)
ax1.set_xlim(x_limits)

# Create the second histogram
ax2.hist(pa_discs, bins=bins, edgecolor='black')
ax2.set_title('Post Annealing')
ax2.set_xlabel('Disconnectivity')
ax2.set_ylabel('Frequency')
ax2.set_ylim(y_limits)
ax2.set_xlim(x_limits)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show plot
plt.show()
```
